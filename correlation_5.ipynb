```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#read target data csv for y values
target_df=pd.read_csv('target.csv')
target_df.head()
```

```python
#read snotel data to determine stations with highest coorelation to target value
snotel_df=pd.read_csv('snotel.csv')
snotel_df.head()
```

```python
stations_df=pd.read_csv('stations.csv')
stations_df = stations_df.rename(columns={'Station Name':'Station'})
stations_df.head()
```

```python
station_list=snotel_df.columns.str.split("(").str[0]
station_list
```

```python
snotel_df.set_axis(station_list, axis='columns', inplace=True)
snotel_df = snotel_df.rename(columns = lambda x : str(x)[:-1])
snotel_df = snotel_df.rename(columns={'Dat':'Date'})
snotel_df.head()
```

```python
#correlation to 3 day forecast
merged_df = pd.merge(target_df[['Date','binary_5']],snotel_df.iloc[:,:],how='left', on ='Date')
merged_df.head()
```

```python

corr_df = merged_df.corr()[['binary_5']].sort_values('binary_5',ascending = False)

corr_df.head()
```

```python
corr_df=corr_df.reset_index()
```

```python
corr_df.head()
```

```python
corr_df = corr_df.rename(columns={"index": "Station"})
```

```python
#plot stations with correlation
```

```python
plot_df = pd.merge(corr_df,stations_df,how = 'left',on = 'Station')
plot_df = plot_df.drop([0])
```

```python
import plotly.express as px

fig = px.scatter_mapbox(plot_df, lat="Latitude", lon="Longitude", hover_name='Station' ,   color="binary_5", 
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=3.5)
fig.update_layout(mapbox_style="open-street-map")

fig.show()
```

```python
fig.write_html("day5.html")
```

```python
#drop stations with missing values
merged_df= merged_df.dropna(axis='columns')
```
